```python
# Drug Side Effect Profiling - Pharmacovigilance Analytics Demo
# -------------------------------------------------------------
# This notebook simulates pharmacovigilance analysis using dummy data.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------
# 1. Create a Dummy Dataset
# ---------------------------
np.random.seed(42)

drugs = ["Drug_A", "Drug_B", "Drug_C", "Drug_D"]
side_effects = ["Nausea", "Headache", "Dizziness", "Fatigue", "Insomnia", "Rash", "Anxiety"]

data = {
    "Drug": np.random.choice(drugs, 200),
    "Side_Effect": np.random.choice(side_effects, 200),
    "Cases": np.random.randint(5, 100, 200),
    "Severity": np.random.choice(["Mild", "Moderate", "Severe"], 200, p=[0.5, 0.3, 0.2])
}

df = pd.DataFrame(data)
df.head()
```

```python
# ---------------------------
# 2. Basic Summary Statistics
# ---------------------------
print("Dataset Overview:")
print(df.describe(include="all"))

print("\nTop 5 Most Common Side Effects:")
print(df["Side_Effect"].value_counts().head())
```

```python
# ---------------------------
# 3. Frequency Analysis
# ---------------------------
side_effect_freq = df.groupby("Side_Effect")["Cases"].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=side_effect_freq.index, y=side_effect_freq.values, palette="viridis")
plt.title("Total Reported Cases per Side Effect")
plt.ylabel("Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------------------------
# 4. Drug vs Side Effect Heatmap
# ---------------------------
heatmap_data = df.pivot_table(values="Cases", index="Drug", columns="Side_Effect", aggfunc="sum", fill_value=0)

plt.figure(figsize=(8,6))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlOrBr")
plt.title("Drug vs Side Effect Cases")
plt.show()
```

```python
# ---------------------------
# 5. Severity Distribution
# ---------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Severity", data=df, palette="Set2")
plt.title("Distribution of Adverse Event Severity")
plt.show()
```

```python
# ---------------------------
# 6. Export Results
# ---------------------------
df.to_csv("drug_side_effect_analysis.csv", index=False)
print("Analysis exported to drug_side_effect_analysis.csv")
```
